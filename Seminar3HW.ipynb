1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
class Delta:
    def __init__(self, col1, col2):
        self.col1 = col1
        self.col2 = col2
    
    def create_delta_renovated(self, df):
        df['delta_renovated'] = df.apply(lambda x: x[self.col1] - x[self.col2] if x[self.col1] != 0 else 0, axis=1)
        return df


delta = Delta('yr_renovated', 'yr_built')
df = delta.create_delta_renovated(df)
display(df['delta_renovated'].unique())
df.head()
```

1.3 Создайте признаки года продажи, месяца продажи

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%dT%H%M%S')
df['year_sold'] = df['date'].dt.year
df['month_sold'] = df['date'].dt.month_name()
display(f"Уникальные значения месяца продажи: {df['month_sold'].unique()}")
display(f"Уникальные значения года продажи: {df['year_sold'].unique()}")
df.head()
```

1.4 Удалите признаки date, zipcode, lat, long

```python
class Droppy:
    def __init__(self, df, columns_to_drop):
        self.df = df
        self.columns_to_drop = columns_to_drop
    
    def drop_columns(self):
        self.df = self.df.drop(self.columns_to_drop, axis=1)
        return self.df


droppy = Droppy(df, ['date', 'zipcode', 'lat', 'long'])
df = droppy.drop_columns()
df.head()
```

```python
import pandas as pd 

df = pd.read_csv("kc_house_data.csv")
df.head()
# 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
#Отсортируйте от меньшей стоимости к большей
df.groupby('bedrooms').agg({"price": "mean"}).sort_values(by='price', ascending=True)
```

```python
# 3.2. Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома
df.pivot_table(index='condition', values='price', aggfunc=['min', 'mean', 'max'], margins=True, margins_name='Total')
```

```python
# 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида
pd.crosstab(df['waterfront'], df['view'])
```

```python
# 3.4 Каких домов в зависимости от этажности и количества спален больше?
pd.pivot_table(df, index='floors', columns='bedrooms', aggfunc='count', values='id')
```

```python
# 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
pd.pivot_table(df, index='condition', columns='grade', values='price', aggfunc='median')
```
